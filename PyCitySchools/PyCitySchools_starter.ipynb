```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
district_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Calculate the total number of schools
school_count = district_data_complete['school_name'].value_counts()
total_schools = school_count.count()
total_schools
```

```python
#Calculate the total number of students
total_students = district_data_complete['student_name'].count()
total_students
```

```python
#Calculate the total Budget
school_budgets = district_data_complete['budget'].unique()
total_budget = school_budgets.sum()
total_budget
```

```python
#Calculate the average math score
avg_math = district_data_complete['math_score'].mean()
avg_math
```

```python
#Calculate the average reading score
avg_reading = district_data_complete['reading_score'].mean()
avg_reading
```

```python
#Calculate the percentage of students with a passing math score (70 or greater)
passing_math = district_data_complete.loc[district_data_complete['math_score'] >= 70]
percent_passing_math = (passing_math['student_name'].count() / total_students) * 100
percent_passing_math
```

```python
#Calculate the percentage of students with a passing reading score (70 or greater)
passing_reading = district_data_complete.loc[district_data_complete['reading_score'] >= 70]
percent_passing_reading = (passing_reading['student_name'].count() / total_students) * 100
percent_passing_reading
```

```python
#Calculate the percentage of students who passed math and reading (% Overall Passing)
#use &
overall_passing = district_data_complete.loc[(district_data_complete['math_score'] >= 70)&(district_data_complete['reading_score'] >= 70)]
percent_overall_passing = (overall_passing['student_name'].count() / total_students) * 100
percent_overall_passing
```

```python
#Create a dataframe to hold the above results 
raw_data_info = {
    "Total Schools": [total_schools],
    "Total Students": [total_students],
    "Total Budget": [total_budget],
    "Average Math Score": [avg_math],
    "Average Reading Score": [avg_reading],
    "% Passing Math" : [percent_passing_math],
    "% Passing Reading" : [percent_passing_reading],
    "% Overall Passing": [percent_overall_passing]
}
district_summary = pd.DataFrame(raw_data_info, columns=["Total Schools", 
                                               "Total Students", 
                                               "Total Budget", 
                                               "Average Math Score",
                                               "Average Reading Score",
                                               "% Passing Math",
                                               "% Passing Reading",
                                               "% Overall Passing"])

district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)
district_summary["Average Math Score"] = district_summary["Average Math Score"].map("{:.2f}".format)
district_summary["Average Reading Score"] = district_summary["Average Reading Score"].map("{:.2f}".format)
district_summary["% Passing Math"] = district_summary["% Passing Math"].map("%{:.2f}".format)
district_summary["% Passing Reading"] = district_summary["% Passing Reading"].map("%{:.2f}".format)
district_summary["% Overall Passing"] = district_summary["% Overall Passing"].map("%{:.2f}".format)
district_summary
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
#creat array of school names
school_name = district_data_complete["school_name"].unique()
new_school_summary = pd.DataFrame([])

for value in school_name:

    #creat new DF based on school name identified by "school_name" and re-index based on "school_name"
    school = district_data_complete.loc[district_data_complete["school_name"]==value]
    school_ind = school.set_index("school_name")
    
    #Get school type data
    school_type = school_ind.iloc[0, 7]
    
    #perform calculations using the new DF
    total_students_sch = school_ind["Student ID"].count()
    total_budget_sch = school_ind.iloc[0, 9]
    per_student_budget = total_budget / total_students
    avg_math_sch = school_ind["math_score"].mean()
    avg_reading_sch = school_ind["reading_score"].mean()
    passing_math_sch = school_ind.loc[school_ind['math_score'] >= 70]
    percent_passing_math_sch = (passing_math_sch['student_name'].count() / total_students_sch) * 100
    passing_reading_sch = school_ind.loc[school_ind['reading_score'] >= 70]
    percent_passing_reading_sch = (passing_reading_sch['student_name'].count() / total_students_sch) * 100
    overall_passing_sch = school_ind.loc[(school_ind['math_score'] >= 70)&(school_ind['reading_score'] >= 70)]
    percent_overall_passing_sch = (overall_passing_sch['student_name'].count() / total_students_sch) * 100
    
    #creat new DF with all calculations based on school
    raw_data_info_sch = {
        "School Name": [value],
        "School Type": [school_type],
        "Total Students": [total_students_sch],
        "Total School Budget": [total_budget_sch],
        "Average Math Score": [avg_math_sch],
        "Average Reading Score": [avg_reading_sch],
        "% Passing Math" : [percent_passing_math_sch],
        "% Passing Reading" : [percent_passing_reading_sch],
        "% Overall Passing": [percent_overall_passing_sch]
    }
    school_summary = pd.DataFrame(raw_data_info_sch, columns=["School Name",
                                               "School Type", 
                                               "Total Students", 
                                               "Total School Budget", 
                                               "Average Math Score",
                                               "Average Reading Score",
                                               "% Passing Math",
                                               "% Passing Reading",
                                               "% Overall Passing"])
    #creat new dataframe and concact each new DF together to creat a summary table including all schools
    df = [new_school_summary, school_summary]
    new_school_summary = pd.concat(df)

#sorts DF alphabetically by school name
sorted_school_summary = new_school_summary.sort_values("School Name")

#re-indexes DF by school name
district_school_summary = sorted_school_summary.set_index("School Name")

district_school_summary["Total Students"] = district_school_summary["Total Students"].astype(int).map("{:,}".format)
district_school_summary["Total School Budget"] = district_school_summary["Total School Budget"].astype(int).map("${:,.2f}".format)
district_school_summary["Average Math Score"] = district_school_summary["Average Math Score"].map("{:.2f}".format)
district_school_summary["Average Reading Score"] = district_school_summary["Average Reading Score"].map("{:.2f}".format)

district_school_summary
```

```python
#merges DF's from above 4 cells into one DF
mbg_1_merge = pd.merge(nine_grades_school, ten_grades_school, on="school_name")
mbg_1 = mbg_1_merge.rename(columns={"math_score_x":"9th","math_score_y":"10th"})
mbg_2_merge = pd.merge(mbg_1, eleven_grades_school, on="school_name")
mbg_2 = mbg_2_merge.rename(columns={"math_score":"11th"})
mbg_3_merge = pd.merge(mbg_2, twelve_grades_school, on="school_name")
math_by_grade = mbg_3_merge.rename(columns={"math_score":"12th"})
math_by_grade
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
#creates column of "spending per student" and drops unnecessary columns
sorted_school_summary["Spending Per Student"] = sorted_school_summary["Total School Budget"].astype(float)/sorted_school_summary["Total Students"].astype(int)
scores_spending_clean = sorted_school_summary.drop(columns=["School Type", "School Name", "Total Students", "Total School Budget"])
```

```python
#creats bins and group names. Sorts data into bins.
bins = [0, 584.9,629.9,644.9,680]
group_names = ["<$585", "$585-630", "$630-645", "$645-680"]
scores_spending_clean["Spending Ranges(Per Student)"]=pd.cut(scores_spending_clean["Spending Per Student"], bins, labels=group_names)
```

```python
#groups by bins and removes final unnecessary column. 
df = scores_spending_clean.groupby("Spending Ranges(Per Student)").mean()
scores_by_spending =df.drop(columns = ["Spending Per Student"])

#Formats data to 2 decimal places.
scores_by_spending["Average Math Score"] = scores_by_spending["Average Math Score"].map("{:.2f}".format)
scores_by_spending["Average Reading Score"] = scores_by_spending["Average Reading Score"].map("{:.2f}".format)
scores_by_spending["% Passing Math"] = scores_by_spending["% Passing Math"].map("{:.2f}".format)
scores_by_spending["% Passing Reading"] = scores_by_spending["% Passing Reading"].map("{:.2f}".format)
scores_by_spending["% Overall Passing"] = scores_by_spending["% Overall Passing"].map("{:.2f}".format)
scores_by_spending
```
